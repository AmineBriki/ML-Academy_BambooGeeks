```python
# 1:1
# 2: A,B,C,2
# 3: D,E,F,3
# 4: G,H,I,4
# 5: J,K,L,5
# 6: M,N,O,6
# 7: P,Q,R,S,7
# 8: T,U,V,8
# 9: W,X,Y,Z,9
# *: *
# 0: [space],0
# #: #
```

```python
x = input("Enter Your Phrase: ")
x=x.lower()
y=list(x)
t=0
dct = {}
st0=['!', '"', '#', '$', '%', '&', '\'', '(', ')', '*', '+', ',', '-', '.', '/', ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`',
     '{', '|', '}', '~'] 
st1 = ['1', 'a', 'd', 'g', 'j', 'm', 'p', 't', 'w', '*', ' ', '#']
st2 = ['b', 'e', 'h', 'k', 'n', 'q', 'u', 'x', '0']
st3 = ['c', 'f', 'i', 'l', 'o', 'r', 'v', 'y']
st4 = ['2', '3', '4', '5', '6', 's', '8', 'z']
st5 = ['7', '9']
for a in st0:
  dct[a] = 0
for a in st1:
  dct[a] = 1
for a in st2:
  dct[a] = 2
for a in st3:
  dct[a] = 3
for a in st4:
  dct[a] = 4
for a in st5:
  dct[a] = 5
for i in y:
  t=t+dct[i]
print("The amount of button presses:",t)
```
